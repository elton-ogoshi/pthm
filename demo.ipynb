```python

import pandas as pd
from pymatgen.core import Element
from pthm.core import PeriodicTableHeatMap


# Create a dataframe for the periodic table
heatmap_data = []

for element in Element:
    if element.row == 6 and element.group  == 3:
        group_index = element.Z - 53
        row_index = 9
    elif element.row == 7 and element.group  == 3:
        group_index = element.Z - 85
        row_index = 10
    else:
        group_index = element.group
        row_index = element.row

    heatmap_data.append({
        'row': row_index,
        'group': group_index,
        'property': element.atomic_radius, # Adapt the code here for a custom property
        'property_str': str(element.atomic_radius), # Adapt the code here for a custom property
        'symbol': element.symbol,
        'number': element.Z  
    })
    

df = pd.DataFrame(heatmap_data)
pt_hm = PeriodicTableHeatMap(df)
pt_hm.plot()
```

```python
pt_hm.save_fig('atomic_radius_heatmap.pdf')
```
