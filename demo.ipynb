This notebook shows how to use the `PeriodicTableHeatMap` class to plot a specific elemental property from a .csv file (in this example, atomic radius)

```python
import pandas as pd
from pthm.core import PeriodicTableHeatMap

# The dataframe should contain a column called 'elements'.
df = pd.read_csv('etc/atomic_radius.csv')
df.head()
```

```python
pt_hm = PeriodicTableHeatMap(df)
```

For a clean look heatmap

```python
pt_hm.plot(property_col='atomic_radius', 
           show_number=False,
           show_values=False)
```

For a heatmap with all numeric values of the property displayed. It also displays the atomic number by default.

```python
# Clean look
pt_hm.plot(property_col='atomic_radius', 
           legend_title='Atomic\nRadius\n (Å)\n \n')
```

```python
pt_hm.save_fig('etc/atomic_radius_heatmap.pdf')
```
